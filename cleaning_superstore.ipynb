## Clean column names

```python
df.columns = df.columns.str.lower().str.replace(' ', '_').str.replace('-', '_')
```

```python
df.columns
```

## Clean up columns

```python
#set date objects
d_cols=['order_date', 'ship_date']
for d in d_cols:
    df[d]=pd.to_datetime(df[d], format='%d/%m/%Y')
```

```python
df.info()
```

```python
#looking at the null postal codes
df[df['postal_code'].isna()]
```

```python
# there are no other entries for the same customers in Burlington
# so we will replace missing zipcodes with main Burlington zip 05401
df['postal_code']=df['postal_code'].astype(str)
```

```python
df['postal_code']=df['postal_code'].str.replace('nan', '05401')
```

```python
df['postal_code']=df['postal_code'].str.slice(stop=5)
```

```python
df.head()
```

```python
#round sales figures to two places
df['sales']=df['sales'].round(2)
```

```python
#check category and sub-category for consistency
df['category'].unique()
```

```python
for c in list(df['category'].unique()):
    df_temp = df[df['category'] == c]
    print(c)
    print(df_temp['sub_category'].unique())
```

```python
#save cleaned dataset
df.to_csv('data/superstore_clean.csv', index=False)
```
